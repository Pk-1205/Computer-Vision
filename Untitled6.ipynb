```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def detect_edges(image_path):
    # Load image
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError("Image not found!")

    # Resize for display (optional)
    resized = cv2.resize(img, (600, 600))

    # Convert to grayscale
    gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)

    # Blur
    gray = cv2.GaussianBlur(gray, (5, 5), 0)

    # Canny edge detection
    edges = cv2.Canny(gray, 50, 150)

    # Morph close â†’ to join broken edges
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    edges_closed = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)

    return resized, edges, edges_closed


# ---- RUN ----

path = "/content/simple.jpeg"
input_img, canny_edges, final_edges = detect_edges(path)

print(" Input Image")
cv2_imshow(input_img)

print(" Canny Edges")
cv2_imshow(canny_edges)

print(" Final Edges (After MORPH_CLOSE)")
cv2_imshow(final_edges)
```

```python
import cv2
import numpy as np
import imutils
from google.colab.patches import cv2_imshow


# Helper Functions

def order_points(pts):
    rect = np.zeros((4, 2), dtype="float32")
    s = pts.sum(axis=1)
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]
    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]
    rect[3] = pts[np.argmax(diff)]
    return rect


def four_point_transform(image, pts):
    rect = order_points(pts)
    (tl, tr, br, bl) = rect

    widthA = np.linalg.norm(br - bl)
    widthB = np.linalg.norm(tr - tl)
    maxWidth = int(max(widthA, widthB))

    heightA = np.linalg.norm(tr - br)
    heightB = np.linalg.norm(tl - bl)
    maxHeight = int(max(heightA, heightB))

    dst = np.array([[0, 0],
                    [maxWidth - 1, 0],
                    [maxWidth - 1, maxHeight - 1],
                    [0, maxHeight - 1]], dtype="float32")

    M = cv2.getPerspectiveTransform(rect, dst)
    warped = cv2.warpPerspective(image, M, (maxWidth, maxHeight))
    return warped


def find_document_contour(edged):
    cnts = cv2.findContours(edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    cnts = sorted(cnts, key=cv2.contourArea, reverse=True)[:10]

    for c in cnts:
        peri = cv2.arcLength(c, True)
        approx = cv2.approxPolyDP(c, 0.02 * peri, True)
        if len(approx) == 4:
            return approx.reshape(4, 2)
    return None


# Main Scanner Function

def detect_document_and_edges(image_path):
    orig = cv2.imread(image_path)
    if orig is None:
        raise FileNotFoundError("Image not found!")

    # Resize for display
    ratio = orig.shape[0] / 500.0
    image = imutils.resize(orig, height=500)

    # Convert to gray
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray_blur = cv2.GaussianBlur(gray, (5, 5), 0)

    #  EDGE DETECTION
    edges = cv2.Canny(gray_blur, 50, 150)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    edges_closed = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)

    # ---- DOCUMENT DETECTION ----
    doc_cnt = find_document_contour(edges_closed)

    if doc_cnt is None:
        print(" No document found. Showing only edges.")
        return orig, edges_closed, None, None

    doc_cnt_scaled = doc_cnt * ratio
    warped = four_point_transform(orig, doc_cnt_scaled)

    # Process B/W scan
    warped_gray = cv2.cvtColor(warped, cv2.COLOR_BGR2GRAY)
    processed = cv2.adaptiveThreshold(
        warped_gray, 255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY, 11, 2
    )

    return orig, edges_closed, warped, processed


# RUN

path = "/content/simple.jpeg"

input_img, edge_img, doc_img, bw_img = detect_document_and_edges(path)

# SHOW RESULTS

print(" Input Image")
cv2_imshow(imutils.resize(input_img, width=600))

print(" Edge Image")
cv2_imshow(imutils.resize(edge_img, width=600))

if doc_img is not None:
    print(" Detected Document (Color Scan)")
    cv2_imshow(imutils.resize(doc_img, width=600))

    print(" B/W Processed Scan")
    cv2_imshow(imutils.resize(bw_img, width=600))
else:
    print(" No document found, so document scan not displayed.")
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def detect_edges_clean(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError("Image not found!")

    img_resized = cv2.resize(img, (600, 600))
    gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)

    # 1) Boost contrast â†’ helps canny find edges
    gray = cv2.equalizeHist(gray)

    # 2) Light noise removal (NOT aggressive)
    gray_blur = cv2.GaussianBlur(gray, (3, 3), 0)

    # 3) Lower-threshold Canny (keeps more edges)
    canny_edges = cv2.Canny(gray_blur, 30, 120)

    # 4) VERY small open â†’ remove tiny noise dots
    kernel_open = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 2))
    noise_removed = cv2.morphologyEx(canny_edges, cv2.MORPH_OPEN, kernel_open, iterations=1)

    # 5) Strong close â†’ join document edges
    kernel_close = cv2.getStructuringElement(cv2.MORPH_RECT, (7, 7))
    morph_edges = cv2.morphologyEx(noise_removed, cv2.MORPH_CLOSE, kernel_close)

    return img_resized, canny_edges, noise_removed, morph_edges



path = "/content/download (1).jpeg"
input_img, canny_img, noise_removed_img, final_edges = detect_edges_clean(path)

print(" Input Image")
cv2_imshow(input_img)

print(" Canny Edge")
cv2_imshow(canny_img)

print(" Noise Removed")
cv2_imshow(noise_removed_img)

print(" Final Clean Edges (Morph Close)")
cv2_imshow(final_edges)
```
